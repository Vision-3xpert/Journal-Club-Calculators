```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
PP = 0
PL = 0

# Define the outcomes of the two clinical trials
# (successes, failures)
prior_n= 100
prior_ev = 10
likelihood_n = 200
likelihood_ev = 15


prior_trial = (prior_ev, prior_n - prior_ev)  # (successes, failures) in the prior trial
posterior_trial = (likelihood_ev,likelihood_n-likelihood_ev)  # (successes, failures) in the new trial

# Define the prior distribution using the prior trial outcomes
prior_distribution = beta(prior_trial[0] + 1, prior_trial[1] + 1)

# Define the likelihood using the new trial outcomes
likelihood = beta(posterior_trial[0] + 1, posterior_trial[1] + 1)

# Update the prior distribution using the likelihood to get the posterior
posterior_distribution = beta(
    prior_trial[0] + posterior_trial[0] + 1,
    prior_trial[1] + posterior_trial[1] + 1)

# Plot the prior, likelihood, and posterior distributions
x = np.linspace(0, 1, 1000)
plt.plot(x, prior_distribution.pdf(x), label='Prior', linestyle=':')
plt.plot(x, likelihood.pdf(x), label='Likelihood',color = "r", linestyle='--')
###############3plt.plot(x, posterior_distribution.pdf(x), label = "posterior",color="g", linestyle = "-")




plt.text(.6, 14, 'Prior = ::::', color = "b",fontsize=12)
plt.text(.6, 13, 'Likelihood = -------',color = "r", fontsize=12)
########################33plt.text(.6, 12, 'Posterior = ______', color = "g",fontsize=12)
plt.show()
```
